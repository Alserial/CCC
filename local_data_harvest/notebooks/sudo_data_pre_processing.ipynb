```python
for i in range(df.shape[1]):
    non_null_count = df.iloc[i].notna().sum() # 统计某个feature有数值的量
    if (non_null_count < 10): # 数值比例少于10%
        print('not nan num in col ', i, ': ', non_null_count)
```

```python
zeros_count = (df == 0).sum()
n = 0
deleted_lst = []
for index, value in zeros_count.items():
    if 'tot' not in index and 'sa2' not in index:
        deleted_lst.append(index)
        print(index)
        n+=1
n
```

```python
age_lst = []
m = 0
for index, value in zeros_count.items():
    if 'per_age' in index and 'age_tot' not in index:
        age_lst.append(index)
        print(index)
        m+=1
m
```

```python
df_deleted = df.drop(columns=deleted_lst)
```

```python
df_deleted = df_deleted.drop(columns=age_lst)
```
